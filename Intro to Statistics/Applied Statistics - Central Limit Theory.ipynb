```python
from numpy.random import seed
from numpy.random import randint
from numpy import mean

seed(1)

rolls = randint(1, 7, 50) # rolling a die 50 times 
print(rolls) # each number shows the number that showed up randomly
print(mean(rolls))  # the mean of 50 number that apperaed during each role is equal to 3.44 
```

Caculating the Mean

(1 + 2 + 3 + 4 + 5 + 6) / 6 or 21 / 6 -> We expect that the mean of rolling a die would be 21/6 or 3.5
- As you can see the mean of rollinga die 50 times is 3.44 (our sample) which is close but not equal to theoritical mean(population). 
yet if we increase the sample size we became closer and closer to 3.5. 

```python
seed(1)

rolls = randint(1, 7, 1000) # rolling a die 50 times 
print(rolls) # each number shows the number that showed up randomly
print(mean(rolls))  # the mean of 1000 number that apperaed during each role is equal to 3.49 -> closer to population mean
```

```python
from numpy.random import seed
from numpy.random import randint
from numpy import mean
from matplotlib import pyplot

seed(1)

means = [mean(randint(1, 7, 50)) for _ in range(1000)] # mean of 1000 sample mean -> According to central limit theorom as much as the number of samples increases we will be closer to normal distribution

pyplot.hist(means)
pyplot.show()
```
