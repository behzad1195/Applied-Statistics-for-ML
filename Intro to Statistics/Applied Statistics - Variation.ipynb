```python
from numpy import arange 
from matplotlib import pyplot 
from scipy.stats import norm  

x_axis = arange(-3, 3, 0.001) 

pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.5))  #lower variance has narrowed bell-shapes (blue curve) / std = 0.5
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1)) #higher variance are more spread around the mean hence wide bell-shape (orange curve) / std = 1 
pyplot.show()
```

```python
from numpy.random import seed 
from numpy.random import randn 
from numpy import var

seed(1)
 
data = 5 * randn(10000) + 50 

result = var(data)
print('Variance: %.3f' % result)
```

```python
from numpy.random import seed 
from numpy.random import randn 
from numpy import std

seed(1)
 
data = 5 * randn(10000) + 50 

result = std(data) # STD is a squared root of Variance
print('Standard Deviation: %.3f' % result)
```

```python
# STD can be calculated for any ypes of distribution but if we deal with non-Guassian distribution then we need to first find out what is the type of our distribution and then used the relevant std formula.
```
