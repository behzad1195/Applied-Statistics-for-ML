```python
# pdf and cdf for a normal distribution
from scipy.stats import norm
from matplotlib import pyplot
# define distribution parameters
mu = 50 # The mean
sigma = 5 # The STD
# create distribution
dist = norm(mu, sigma) 
# plot pdf
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values] # pdf probability distribution function
pyplot.plot(values, probabilities)
pyplot.show()
```

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm # norm for normal distribution

sample_space = arange(-5, 5, 0.001) # -5 and 5 for the observation range / 0.001 the lowest difference betwee each observation

cdf = norm.cdf(sample_space) # cdf cumulative distribution function

pyplot.plot(sample_space, cdf)
pyplot.show()
```

```python
# calculate the values that define the middle 95%
from scipy.stats import norm

mu = 50
sigma = 5

dist = norm(mu, sigma)
low_end = dist.ppf(0.025) # in a normal distribution 95% of our data are located beween -2STD and 2STD or between pdf of 0.025 and pdf of 0.975 which their difference is 0.95 or 95 %
high_end = dist.ppf(0.975)
print('Middle 95%% between %.1f and %.1f' % (low_end, high_end))
```

We are expecting that the low and high end would be 40 (mean (50) - 2*STD (5)) and 60 (mean (50) + 2*STD (5)). So our random sample also have very close low and high end with what we expected
