```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2 # scipy stands for scientific python 

sample_space = arange(0, 50, 0.01)
dof = 20 # degrees of freedom -> As you can see for Chi-squared distribution similar to T-distribution we don't need STD and Mean the only parametr is dof

pdf = chi2.pdf(sample_space, dof)

pyplot.plot(sample_space, pdf)
pyplot.show()
```

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2

sample_space = arange(0, 50, 0.01)
dof = 20

cdf = chi2.cdf(sample_space, dof)

pyplot.plot(sample_space, cdf)
pyplot.show()
```
