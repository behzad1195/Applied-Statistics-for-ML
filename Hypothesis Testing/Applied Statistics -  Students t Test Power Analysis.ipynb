```python
# estimate sample size via power analysis
# T-test power analysis is for comapring the mean of two guassian distributed variables 
# H0 for the test is that the means of two variables are the same (no difference between samples). The test will provide a p-value that allow us to test he H0 hypothesis
from statsmodels.stats.power import TTestIndPower

# parameters for power analysis
effect = 0.8
alpha = 0.05 # significance level
power = 0.8  # power level for preventing large number of Type II error (False Negative)

# perform power analysis
analysis = TTestIndPower()
result = analysis.solve_power(effect, power=power, nobs1=None, ratio=1.0, alpha=alpha)
# nobs1 is number of observations of sample 1. The number of observations of sample two is ratio times the size of sample 1, i.e. nobs2 = nobs1 * ratio

print('Sample Size: %.3f' % result) # by giving pre-defined values to the 3 parameters it gives you back the minimum of sample size
```

```python
# calculate power curves for varying sample and effect size
from numpy import array
from matplotlib import pyplot
from statsmodels.stats.power import TTestIndPower

# parameters for power analysis
effect_sizes = array([0.2, 0.5, 0.8])
sample_sizes = array(range(5, 100))

# calculate power curves from multiple power analyses
analysis = TTestIndPower()
analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes)
pyplot.show()
```
